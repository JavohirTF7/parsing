```python
from PyPDF2 import PdfFileReader

def pdf_to_text(path):
    pdf = PdfFileReader(open(path, "rb"))
    text = ""
    for page in range(pdf.getNumPages()):
        text += pdf.getPage(page).extractText()
    return text
```

```python
import PyPDF2

def pdfdan_matn_olish(fayl_nomi):
    with open(fayl_nomi, 'rb') as fayl:
        pdf = PyPDF2.PdfReader(fayl)
        matn = ''
        for sahifa in pdf.pages:
            matn += sahifa.extract_text()
        return matn

```

```python
import torch
from transformers import T5Tokenizer, T5ForConditionalGeneration

# T5 modelini yuklab olish
model_name = "t5-base"
tokenizer = T5Tokenizer.from_pretrained("t5-small")
model = T5ForConditionalGeneration.from_pretrained("t5-small")

# Tarjima uchun ma'lumotni tayyorlash
input_text = "Enter your English text here."

# Matnni kodga o'tkazish
input_ids = tokenizer.encode(input_text, return_tensors="pt")

# Tarjima jarayonini amalga oshirish
output = model.generate(input_ids=input_ids, max_length=100, num_beams=5, early_stopping=True)

# Tarjima natijasini chiqarish
translated_text = tokenizer.decode(output[0], skip_special_tokens=True)

# Tarjima natijasini ekranga chiqarish
print("Tarjima natijasi:", translated_text)
```
