```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
bbc_news_all=pd.read_csv("data/malumotlar/ziyouz/ziyo_uz_links_all(2022.09.22).csv")
bbc_news_all
```

# Linkini list ko'rinishda saqlash

```python
a=bbc_news_all.loc[2500:3000].values.tolist()
a
```

```python
b=[]
for i in a:
    b+=i
print(b)
```

```python

# Saytlar ro'yxati
urls = b

# Barcha matnlar uchun to'plangan ro'yxat
all_matnlar = []

for url in urls:
    # Web sahifani HTML ma'lumotlarini olish
    response = requests.get(url)
    html_content = response.text

    # BeautifulSoup kutubxonasini ishlatib HTML-ni parsing qilish
    soup = BeautifulSoup(html_content, 'html.parser')

    # Matn elementlarini izlash va ro'yxatga qo'shish
    matnlar = soup.find_all('p')  # Misol uchun <p> teglaridagi matnlarni izlaymiz
    all_matnlar.extend(matnlar)

```

# Matnlar faylga saqlanishi

```python
for matn in all_matnlar:
    matn.text
    
with open('New_data/ziyouz/ziyo(2022.09.22)7.txt', 'w', encoding='utf-8') as file:
    for matn in all_matnlar:
        file.write(matn.text + '\n')

print("Matnlar faylga saqlandi: matnlar.txt")
```
